```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
N = [64, 128, 256, 512, 1024]
cpp_int = [0.010, 0.030, 0.121, 0.793, 7.246]
cpp_int_meat = [0.00619, 0.02651, 0.11753, 0.78624, 7.23468]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(N, cpp_int, marker='o', linestyle='-', color='b', label='cpp_int')
plt.plot(N, cpp_int_meat, marker='o', linestyle='--', color='r', label='cpp_int_meat')

plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Performance Comparison for language cpp and datatype int')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()

```
